```python

# Step 1: Import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Step 2: Sample dataset
data = {
    'Age': [22, 25, 47, 52, 46, 56, 55, 60],
    'Income': [25000, 32000, 85000, 90000, 60000, 95000, 75000, 62000],
    'Buy': ['No', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data)

# Step 3: Split into inputs and labels
X = df[['Age', 'Income']]         # Features
y = df['Buy']                     # Labels

# Step 4: Create and train model
model = DecisionTreeClassifier()
model.fit(X, y)

# Step 5: Visualize the tree
plt.figure(figsize=(10,6))
tree.plot_tree(model, feature_names=['Age', 'Income'], class_names=['No', 'Yes'], filled=True)
plt.title("Decision Tree - Buy or Not")
plt.show()

# Step 6: Predict for a new person
# Example: Age=30, Income=40000
prediction = model.predict([[30, 40000]])
print("Will the person buy the product?", prediction[0])
```
